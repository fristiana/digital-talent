### Building Multiple Linear Regression 

#### 1. Visualization

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)
%matplotlib inline

df_ekspor = df[['nilai_ekspor', 'nilai_impor', 'Nilai_IHSG', 'CountEvent06']]
sns.pairplot(df_ekspor, x_vars=['nilai_impor','Nilai_IHSG','CountEvent06'], y_vars='nilai_ekspor', kind='reg')
```

#### 2. SK LearnSetting variables

```python
feature_cols = ['nilai_impor', 'Nilai_IHSG', 'CountEvent06']
x = df_ekspor[feature_cols]
y = df_ekspor['nilai_ekspor']
```

#### 3. Splitting our data

```python
from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

#### 4. Training our models

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(x_train, y_train)
```

#### 5. Interpreting Coefficient

```python
print(linreg.intercept_)
print(linreg.coef_)
zip(feature_cols, linreg.coef_)
```

jadi model yang terbentuk y = 3239887737.729148 + (7.61671328e-01)*nilai_impor + (1.92007591e+05)*Nilai_IHSG - (5.56383129e+05)*CountEvent06
artinya nilai impor dan nilai ihsg berkorelasi positif terhadap nilai ekspor, jika nilai tersebut naik maka ekspor naik namun jika turun maka ekspor turun. Sementara itu CountEvent06 berkorelasi negatif terhadap nilai ekspor, jika jumlah event06 naik maka nilai ekspor turun dan sebaliknya.

#### 6. Making prediction based on our model

```python
y_pred = linreg.predict(x_test)
y_pred
```

#### 7. Model Evaluation

There are three primary metrics used to evaluate linear models. These are: Mean absolute error (MAE), Mean squared error (MSE), or Root mean squared error (RMSE).

MAE: The easiest to understand. Represents average error

MSE: Similar to MAE but noise is exaggerated and larger errors are punished. It is harder to interpret than MAE as its not in base units, however, it is generally more popular.

RMSE: Most popular metric, similar to MSE, however, the result is square rooted to make it more interpretable as its in base units. It is recommended that RMSE be used as the primary metric to interpret your model.

Below, you can see how to calculate each metric. All of them require two lists as parameters, with one being your predicted values and the other being the true values



```python
from sklearn import metrics
#print result of MAE
print(metrics.mean_absolute_error(y_test, y_pred))

#print result of MSE
print(metrics.mean_squared_error(y_test, y_pred))

#print result of RMSE
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

### Regresi dengan data yang distandarisasi

```python
from sklearn.preprocessing import MinMaxScaler

minmax_scaler = MinMaxScaler()
standard_df_ekspor = minmax_scaler.fit_transform(df_ekspor)
standard_df_ekspor
feature_cols = ['nilai_ekspor','nilai_impor', 'Nilai_IHSG', 'CountEvent06']
df_standard = pd.DataFrame(standard_df_ekspor.T, feature_cols)
df_standard = df_standard.T
# df_standard['nilai_impor']
x = df_standard[['nilai_impor', 'Nilai_IHSG', 'CountEvent06']]
y = df_standard['nilai_ekspor']
x_train, x_test, y_train, y_test = train_test_split(x,y)
linreg = LinearRegression()
linreg.fit(x_train, y_train)
```

```python
sns.pairplot(df_standard, x_vars=['nilai_impor','Nilai_IHSG','CountEvent06'], y_vars='nilai_ekspor', kind='reg')
```

```python
print(linreg.intercept_)
print(linreg.coef_)
zip(feature_cols, linreg.coef_)
```

jadi model yang terbentuk y = 0.91130442  - (0.00484814 )*nilai_impor - (0.04584656)*Nilai_IHSG - (0.00875361)*CountEvent06
artinya nilai impor berkorelasi positif terhadap nilai ekspor, jika nilai tersebut naik maka ekspor naik namun jika turun maka ekspor turun. Sementara itu nilai IHSG dan CountEvent06 berkorelasi negatif terhadap nilai ekspor, jika jumlah event06 naik maka nilai ekspor turun dan sebaliknya.

```python
y_pred = linreg.predict(x_test)
y_pred
```

```python
from sklearn import metrics
#print result of MAE
print(metrics.mean_absolute_error(y_test, y_pred))

#print result of MSE
print(metrics.mean_squared_error(y_test, y_pred))

#print result of RMSE
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
