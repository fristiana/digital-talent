```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'group_project'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
sql = "SELECT tahun, AVG(nilai_ekspor) rerataEkspor, AVG(nilai_impor) rerataImpor FROM tabel_eksim GROUP BY tahun"
df = run(sql)
df
```

```python
# df["tahun"].values
df_toplot = pd.DataFrame({'rerataEkspor': df['rerataEkspor'].values,'rerataImpor':df['rerataImpor'].values}, index=df["tahun"].values)
# df_toplot
line = df_toplot.plot(kind='line', title ="Trend Ekspor-Impor Indonesia", legend=True, fontsize=12, figsize=[20,10])
line.set_xlabel("Tahun", fontsize=12)
line.set_ylabel("Nilai (US$)", fontsize=12)
```
