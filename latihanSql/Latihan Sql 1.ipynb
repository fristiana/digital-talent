```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

###  1. Prepare a list of offices sorted by country, state, city.

```python
run("SELECT * FROM `offices` ORDER BY country, state, city")
```

###  2. How many employees are there in the company?

```python
run("SELECT COUNT(employeeNumber) as jmlPegawai FROM `employees`")
```

### 3. What is the total of payments received?

```python
run("SELECT SUM(amount) as jmlPayment FROM `payments` ")
```

### 4. List the product lines that contain 'Cars'.

```python
run("SELECT * FROM `productlines` WHERE productLine LIKE '%Cars%' ")
```

### 5. Report total payments for October 28, 2004.

```python
run("SELECT SUM(amount) FROM `payments` WHERE paymentDate <= '2004-10-28' ")
```

### 6. Report those payments greater than \$100,000.

```python
run("SELECT * FROM `payments` WHERE amount > 100000.00 ")
```

### 7. List the products in each product line.

```python
listProductline = run("SELECT productline FROM `productlines`")
# len(listProductLine)
i = 0
while i<len(listProductLine):
#     print(i)
    productline = listProductLine['productline'][i]
    print('daftar produk untuk productline:',productline)
    sql = "SELECT productCode, productName FROM `products` WHERE productline = '"+productline+"'"
#     print(sql)
    daftar = run(sql)
    print(daftar)
    print('\n\n')
    i = i+1
```

### 8. How many products in each product line?

```python
run("SELECT productline, COUNT(productCode) as jmlProduk FROM `products` GROUP BY productline")
```

### 9. What is the minimum payment received?

```python
run("SELECT MIN(amount) FROM `payments`")
```

### 10. List all payments greater than twice the average payment.

```python
run("SELECT * FROM payments WHERE amount > (SELECT 2*AVG(amount) as twiceRerata FROM payments)")
```

### 11. What is the average percentage markup of the MSRP on buyPrice?

```python
run("SELECT AVG((MSRP - buyPrice)*100.00/buyPrice) as averagePercentageMarkup FROM `products` ")
```

### 12. How many distinct products does ClassicModels sell?

```python
run("SELECT COUNT(Distinct(productCode)) as jmlJenisProdukBerbeda FROM products WHERE productline like '%classic%'")
```

### 13. Report the name and city of customers who don't have sales representatives?

```python
run("SELECT customerName, city FROM `customers` WHERE salesRepEmployeeNumber is null")
```

### 14. What are the names of executives with VP or Manager in their title? Use the CONCAT function to combine the employee's  first name and last name into a single field for reporting.

```python
run("SELECT CONCAT(firstName,' ',lastName) as name, jobTitle FROM `employees` WHERE jobTitle like '%VP%' OR jobTitle like '%Manager%' ")
```

### 15. Which orders have a value greater than $5,000?

```python
run("SELECT orderNumber, productCode, quantityOrdered, priceEach, (quantityOrdered*priceEach) as value FROM `orderdetails` WHERE (quantityOrdered*priceEach) > 5000 ")
```
