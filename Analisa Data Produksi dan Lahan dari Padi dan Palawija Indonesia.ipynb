Dataset yang digunakan dalam analisa ini adalah dataset yang diambil dari data.go.id yang berisi informasi mengenai produksi, produktivitas, dan luas panen dari Padi dan Palawija di Indonesia dari tahun 1969 - 2014.
Penjelasan mengenai variabel yang ada pada dataset ini adalah sebagai berikut:

### Contoh data asli
berikut ini merupakan set data aslinya diambil beberapa sampel untuk memperlihatkan bentuk dataset awal

```python
import pandas as pd

data = pd.read_csv('processed-data-padi-palawija-1969-2014.csv', delimiter = ',', encoding = "ISO-8859-1")
data.sample(30)
```

### Jenis Indikator dan Jenis Komoditas/Tanaman yang Ada pada Dataset

```python
data.groupby(['indikator', 'jenis_tanaman']).mean()['nilai']
```

### Narasi Output
Dari keluaran tersebut dapat dilihat bahwa ada 2 lembaga sebagai sumber data yaitu: Kementrian Pertanian dan BPS. Yang mana masing-masing lembaga mengeluarkan 3 jenis indikator yaitu: produksi, luas lahan, dan produktivitas untuk 9 jenis tanaman, yaitu: Padi, Padi Ladang, Padi Sawah, Jagung, Kedelai, Kacang Tanah, Kacang Hijau, Ubi Jalar dan Ubi Kayu.

## Perkembangan Luas Panen, Produksi, dan Produktivitas Padi Sawah bersumber BPS

```python
df_prod_padi_bps_series = data.loc[(data['jenis_tanaman'] == 'Padi Sawah') & ((data['indikator'] == ' Produksi - bersumber BPS') | (data['indikator'] == ' Produktivitas - bersumber BPS') | (data['indikator'] == ' Luas Panen - bersumber BPS'))]
df_prod_padi_bps_series
import numpy as np
table = pd.pivot_table(df_prod_padi_bps_series, values='nilai', index=['indikator'],columns=['tahun'], aggfunc=np.sum)
table
```

## Perkembangan Luas Panen, Produksi, dan Produktivitas Padi Sawah bersumber Kementan

```python
df_prod_padi_kementan_series = data.loc[(data['jenis_tanaman'] == 'Padi Sawah') & ((data['indikator'] == 'Produksi - bersumber Kementerian Pertanian') | (data['indikator'] == 'Produktivitas - bersumber Kementerian Pertanian') | (data['indikator'] == 'Luas Panen - bersumber Kementerian Pertanian'))]
df_prod_padi_kementan_series
table_kementan = pd.pivot_table(df_prod_padi_kementan_series, values='nilai', index=['indikator'],columns=['tahun'], aggfunc=np.sum)
table_kementan
```

## Perbandingan Statistik/Sebaran data antara sumber BPS  dengan Kementan untuk Indikator Produksi

```python
df_produksi = data.loc[(data['indikator'] == ' Produksi - bersumber BPS') | (data['indikator'] == 'Produksi - bersumber Kementerian Pertanian')]
table_perbandingan = pd.pivot_table(df_produksi, values='nilai', index=['jenis_tanaman'],columns=['indikator'], aggfunc={'nilai': [min, max, np.median, np.mean]})
table_perbandingan
```

### Narasi Output
Dari hasil perbandingan di atas terutama melihat pada median data, serta series data pada 2 output sebelumnya, dapat dilihat bahwa untuk data yang bersumber dari BPS itu tidak lengkap. Terlihat dari mediannya 0 untuk series data tahun 1969 - 2014. Oleh karena itu untuk selanjutnya analisis mengenai produksi dan lahan Padi Palawija akan dilakukan menggunakan data yang bersumber dari Kementerian Pertanian.
Perihal ketidaklengkapan data dari BPS ini dapat dijadikan bahasan tersendiri. Apakah mungkin dikarenakan pengumpulan data sebelum tahun 2011 adalah murni tanggung jawab Kementan saja sehingga BPS tidak mempunyai datanya atau faktor lainnya.

## Rata-rata Nilai Produktivitas, Produksi dan Luas Lahan menurut Jenis Tanaman/komoditas tahun 2011-2014 

```python
df_kementan20112014 = data.loc[(data['tahun']>2010) & (data['tahun']<2015) & ((data['indikator'] == 'Produktivitas - bersumber Kementerian Pertanian')|(data['indikator'] == 'Produksi - bersumber Kementerian Pertanian')|(data['indikator'] == 'Luas Panen - bersumber Kementerian Pertanian'))]
table_rerata_20112014 = pd.pivot_table(df_kementan20112014, values='nilai', index=['jenis_tanaman'],columns=['indikator'], aggfunc={'nilai': [np.mean]})
table_rerata_20112014
```

```python
# table_rerata_20112014.plot.bar()
bar = table_rerata_20112014.plot(kind='bar', title ="Rata-rata Nilai Produktivitas, Produksi dan Luas Lahan menurut Jenis Tanaman/komoditas tahun 2011-2014", legend=True, fontsize=12, figsize=[20,10])
bar.set_xlabel("Jenis Tanaman", fontsize=12)
bar.set_ylabel("Nilai Rata-rata", fontsize=12)
```

### Komoditas dengan rata-rata luas panen, produksi, atau produktivitas terbesar antara tahun 2011-2014

```python
def getIIndexDictByValue(myDict, myVal):
    for key,value in myDict.items():
        if value == myVal:
            return key

table_rerata2 = table_rerata_20112014.to_dict()
max_luas_panen = max(table_rerata2[('mean',
  'Luas Panen - bersumber Kementerian Pertanian')].values())
kom_max_luas_panen = getIIndexDictByValue(table_rerata2[('mean',
  'Luas Panen - bersumber Kementerian Pertanian')], max_luas_panen)
print('Komoditas dengan rata-rata luas panen terbesar:',kom_max_luas_panen)

max_produksi = max(table_rerata2[('mean',
  'Produksi - bersumber Kementerian Pertanian')].values())
kom_max_produksi = getIIndexDictByValue(table_rerata2[('mean',
  'Produksi - bersumber Kementerian Pertanian')], max_produksi)
print('Komoditas dengan rata-rata produksi terbesar:',kom_max_produksi)

max_produktivitas = max(table_rerata2[('mean',
  'Produktivitas - bersumber Kementerian Pertanian')].values())
kom_max_produktivitas = getIIndexDictByValue(table_rerata2[('mean',
  'Produktivitas - bersumber Kementerian Pertanian')], max_produktivitas)
print('Komoditas dengan rata-rata produktivitas terbesar:',kom_max_produktivitas)
```

### Narasi Output
Berdasarkan Rata-rata Nilai Produktivitas, Produksi dan Luas Lahan menurut Jenis Tanaman/komoditas tahun 2011-2014 dapat dilihat bahwa komoditas dengan produktivitas terbesar adalah Ubi Kayu, artinya Ubi Kayu adalah komoditas strategis yang dikembangkan antara tahun 2011 sampai 2014. Kita perlu menganalisa juga bahwa sebenarnya komoditas Padi memiliki nilai rata-rata produksi tertinggi antara tahun 2011 sampai 2014 berbanding lurus dengan luas panennya yang juga terbesar sehingga jika dihitung nilai produktivitasnya yang merupakan pembagian nilai produksi terhadap luas lahan maka produktivitasnya tidak lah begitu tinggi. Artinya produksi Padi tidak efisien. Seharusnya melihat potensi begitu luasnya lahan Padi bisa dimanfaatkan untuk meningkatkan nilai produksi sehingga produktivitasnya pun meningkat dengan demikian produksi Padi dapat dinyatakan efisien.

Untuk menentukan arah kebijakan dalam bidang pangan tidaklah cukup dengan mengetahui produktivitas atau komoditas strategis dalam satu titik atau satu range acuan waktu tertentu saja. Kita perlu menggali lebih dalam melalui series data dari tahun-tahun sebelumnya untuk melihat trend masing-masing komoditas dan menganalisa potensi komoditas strategis dari trend tersebut.
