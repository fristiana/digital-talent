# Scraping Kompas - Ekonomi
Pada python notebook ini akan dijelaskan mengenai scraping berita dalam ekonomi.kompas.com. Scraping dimulai dari halaman awal, lalu akan menelusuri setiap link, dan akan mengambil berita beserta properties lainnya.

## Inisialisasi urllib.request

```python
import urllib.request

r = urllib.request.urlopen('https://ekonomi.kompas.com/').read()
print(type(r))
print(r[1:1000])
```

## Inisialisasi BeautifulSoup

```python
from bs4 import BeautifulSoup 

soup = BeautifulSoup(r,"lxml")
print(type(soup))
print(soup.prettify()[1:1000])
```

## Dapatkan Semua Link

```python
links = soup.find_all('a')
print(len(links))

for link in links:
#     print(link)
    print('%s: %s'%(link.text.strip(),link["href"]))
```

Link yang benar-benar menuju ke berita adalah link yang mengandung '/read/' dan benar-benar link bukan javascript maupun link '#', sehingga perlu difilter sebagai berikut:

```python
list_key_link = ['/read/']
berita_link = []
for i in list_key_link:
    berita_link = berita_link + [link for link in links if i in str(link) and 'javascript:void(0)' not in str(link) 
               and '#' not in str(link)]
print(len(berita_link))
for link in berita_link:
    print('%s: %s'%(link.text.strip(),link["href"]))
```

## Dapatkan Isi Berita
Di sini akan diambil judul berita, isi berita, dan tanggal berita dipost/dibuat. Dari hasil inspeksi, judul ada pada elemen h1 dengan class read__title. Untuk isi berita ada pada elemen div class read__content. Sedangkan untuk tanggal ada pada elemen div read__time

```python
html = urllib.request.urlopen("https://ekonomi.kompas.com/read/2018/10/24/204900026/pertamina-ingin-beli-minyak-dengan-mata-uang-selain-dollar-as").read()
soup = BeautifulSoup(html, "lxml")

judul = soup.find("h1","read__title").text.strip()
isi_berita = soup.find("div","read__content").text.strip()
tanggal = soup.find("div","read__time").text

tanggal = tanggal.strip().split('-')[1].split('|')[0].strip()
print(tanggal)
print(judul)
print(isi_berita)
```

## Fungsi Scraping Utuh
merupakan fungsi scraping yang utuh untuk scraping berita dari ekonomi.kompas

```python
def getBerita(link):
    html = urllib.request.urlopen(link).read()
    soup = BeautifulSoup(html, "lxml")

    judul = soup.find("h1","read__title").text.strip()
    isi_berita = soup.find("div","read__content").text.strip()
    tanggal = soup.find("div","read__time").text

    tanggal = tanggal.strip().split('-')[1].split('|')[0].strip()
    
    req = urllib.request.Request(
        gambar['src'], 
        data=None, 
        headers={
            'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'
        }
    )
    r = urllib.request.urlopen(req).read()
    i = Image.open(BytesIO(r))
    
    exclude = set(string.punctuation)
    judul = ''.join(str(ch) for ch in judul if ch not in exclude)
    
    nama_file = "images/" + judul + ".jpg"
    i.save(nama_file,'JPEG')
    
    return [judul,isi_berita,tanggal]
```

## Dapatkan semua data dari link berita

```python
# getBerita('https://ekonomi.kompas.com/read/2018/10/24/204900026/pertamina-ingin-beli-minyak-dengan-mata-uang-selain-dollar-as')
berita = []
for link in berita_link:
    print('%s: %s'%(link.text.strip(),link["href"]))
    try:
        berita.append(getBerita(link["href"]))
    except:
        pass
```

## Simpan ke dalam CSV

```python
# print(berita)
import csv

# with open('data_artikel.csv', 'wb') as f:
with open('data_berita.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerows(berita)
```

## Simpan ke dalam XLSX

```python
import pandas as pd
data = pd.DataFrame(berita)
writer = pd.ExcelWriter('data_berita.xlsx')
data.to_excel(writer,'Sheet1',index=False)
writer.save()
```
